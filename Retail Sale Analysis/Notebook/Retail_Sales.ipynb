```python
import pandas as pd
from sqlalchemy import create_engine
import os
os.chdir(r'C:\Users\Muneer Khan\OneDrive\Desktop\Project')
print(os.getcwd())

```

```python
# 1. Load dataset
df = pd.read_csv(r"C:\Users\Muneer Khan\OneDrive\Desktop\Project\retail_sales.csv")
(df.head())
```

```python
# 2. Clean and preprocess
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['Date', 'Sales', 'Profit'])
df['Month'] = df['Date'].dt.month_name()
```

```python
# 3. Derive new metric
df['Profit_Percentage'] = (df['Profit'] / df['Sales'])
```

```python
# 4. Connect to SQLite database
engine = create_engine(r'sqlite:///C:\Users\Muneer Khan\OneDrive\Desktop\Project\retail_sales.db')
```

```python
# 5. Store cleaned data into SQL table
df.to_sql('sales_data', con=engine, if_exists='replace', index=False)

print("✅ Data cleaned and loaded successfully into SQL database!")
```
